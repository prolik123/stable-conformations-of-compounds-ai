```python
import os
import random
import requests
from ase.io import read, write
from ase import Atoms

# Ścieżki folderów
ORIGINAL_DIR = 'original_cifs'
MODIFIED_DIR = 'modified_cifs'

# Utwórz foldery, jeśli nie istnieją
os.makedirs(ORIGINAL_DIR, exist_ok=True)
os.makedirs(MODIFIED_DIR, exist_ok=True)

# Funkcja do pobierania CIF z URL
def download_cif(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)
        return True
    return False

# Funkcja do modyfikacji pozycji atomów
def modify_structure(structure, max_displacement=0.1):
    atoms = structure.copy()
    for atom in atoms:
        dx = random.uniform(-max_displacement, max_displacement)
        dy = random.uniform(-max_displacement, max_displacement)
        dz = random.uniform(-max_displacement, max_displacement)
        atom.position += [dx, dy, dz]
    return atoms

# Wczytaj linki z pliku
with open('COD-selextion.txt', 'r') as file:
    urls = [line.strip() for line in file if line.strip()]

for i, url in enumerate(urls):
    original_filename = os.path.join(ORIGINAL_DIR, f'structure_{i}.cif')
    modified_filename = os.path.join(MODIFIED_DIR, f'structure_{i}_modified.cif')

    print(f"Pobieranie pliku: {url}")
    if download_cif(url, original_filename):
        try:
            structure = read(original_filename)
            modified_structure = modify_structure(structure)

            # Zapisz zmodyfikowaną strukturę
            write(modified_filename, modified_structure)
            print(f"Pobrano i zmodyfikowano strukturę nr {i}")
        except Exception as e:
            print(f"Błąd przy przetwarzaniu pliku {original_filename}: {e}")
    else:
        print(f"Nie udało się pobrać pliku: {url}")
```

```python
import os
from ase.io import read

# Foldery
ORIGINAL_CIF_DIR = "original_cifs"
MODIFIED_CIF_DIR = "modified_cifs"
ORIGINAL_TXT_DIR = "original_txt"
MODIFIED_TXT_DIR = "modified_txt"
MAX_FILES = 500

# Utwórz foldery na dane wyjściowe
os.makedirs(ORIGINAL_TXT_DIR, exist_ok=True)
os.makedirs(MODIFIED_TXT_DIR, exist_ok=True)

# Wybierz wspólne pliki: structure_0.cif ... structure_499.cif
common_names = [f"structure_{i}.cif" for i in range(MAX_FILES)]

for cif_file in common_names:
    basename = os.path.splitext(cif_file)[0]  # np. structure_0

    original_path = os.path.join(ORIGINAL_CIF_DIR, cif_file)
    modified_path = os.path.join(MODIFIED_CIF_DIR, f"{basename}_modified.cif")

    original_out = os.path.join(ORIGINAL_TXT_DIR, f"{basename}.txt")
    modified_out = os.path.join(MODIFIED_TXT_DIR, f"{basename}_modified.txt")

    try:
        # Odczyt i zapis struktury oryginalnej
        atoms_original = read(original_path)
        with open(original_out, "w") as f_out:
            for atom in atoms_original:
                x, y, z = atom.position
                f_out.write(f"{x:.6f} {y:.6f} {z:.6f} {atom.number}\n")
        
        # Odczyt i zapis struktury zmodyfikowanej
        atoms_modified = read(modified_path)
        with open(modified_out, "w") as f_out:
            for atom in atoms_modified:
                x, y, z = atom.position
                f_out.write(f"{x:.6f} {y:.6f} {z:.6f} {atom.number}\n")

        print(f"✔ Przetworzono parę: {basename}")
    
    except Exception as e:
        print(f"❌ Błąd dla {basename}: {e}")
```

```python
import os
from ase.io import read

# Lista nazw twoich plików xyz
xyz_files = ['butan_anti_optymalna.xyz', 'butan_eclipsed_nieoptymalna.xyz', 'butan_gauche_nieoptymalna.xyz']  # <- Podaj swoje nazwy

# Foldery
INPUT_DIR = 'Butan w kilku konformacjach'        # Folder, gdzie masz pliki xyz
OUTPUT_DIR = 'Butan w kilku konformacjach_txt'         # Folder na wyjście .txt

os.makedirs(OUTPUT_DIR, exist_ok=True)

for filename in xyz_files:
    input_path = os.path.join(INPUT_DIR, filename)
    base_name = os.path.splitext(filename)[0]
    output_path = os.path.join(OUTPUT_DIR, f"{base_name}.txt")

    try:
        atoms = read(input_path)
        with open(output_path, "w") as f_out:
            for atom in atoms:
                x, y, z = atom.position
                f_out.write(f"{x:.6f} {y:.6f} {z:.6f} {atom.number}\n")
        print(f"✔ Zapisano {output_path}")
    except Exception as e:
        print(f"❌ Błąd przy przetwarzaniu {filename}: {e}")
```

```python
import numpy as np
import pyvista as pv
import os

# Kolory CPK dla podstawowych pierwiastków
CPK_COLORS = {
    1:  'white',      # H
    6:  'black',      # C
    7:  'blue',       # N
    8:  'red',        # O
    9:  'green',      # F
    16: 'yellow',     # S
}

# Promienie Van der Waalsa (w angstromach)
VDW_RADII = {
    1: 0.31,
    6: 0.76,
    7: 0.71,
    8: 0.66,
    9: 0.57,
    16: 1.05,
}

def load_xyz_file(filepath):
    atoms = []
    with open(filepath, 'r') as file:
        for line in file:
            if not line.strip():
                continue
            parts = line.strip().split()
            if len(parts) < 4:
                continue  # pomiń niepełne linie
            x, y, z, z_atomic = map(float, parts)
            atoms.append((x, y, z, int(z_atomic)))
    return atoms

def visualize_molecule(atoms):
    plotter = pv.Plotter()

    # Wyciągamy współrzędne i centrowanie cząsteczki
    coords = np.array([(x, y, z) for x, y, z, _ in atoms])
    center = coords.mean(axis=0)

    # Rysowanie atomów
    for x, y, z, z_atom in atoms:
        color = CPK_COLORS.get(z_atom, 'gray')
        radius = VDW_RADII.get(z_atom, 0.8)
        position = np.array([x, y, z]) - center
        sphere = pv.Sphere(radius=radius, center=position, theta_resolution=32, phi_resolution=32)
        plotter.add_mesh(sphere, color=color, smooth_shading=True)

    # Stałe osie i ramka
    plotter.show_axes()  # mały lokalny układ osi w rogu
    plotter.show_bounds(grid='front', location='outer', all_edges=True)
    plotter.camera_position = 'xy'  # stała orientacja kamery: patrz z góry

    plotter.show()

if __name__ == "__main__":
    path = r"molecule.txt"
    if not os.path.exists(path):
        print(f"Plik nie istnieje:\n{path}")
    else:
        atoms = load_xyz_file(path)
        visualize_molecule(atoms)
```
